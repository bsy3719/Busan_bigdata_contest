```python
data = pd.read_csv('./subway_main.csv', encoding='euc-kr')
print(data.shape)
data.head()
```

```python
data.head()
```

```python
data.describe()
```

```python
group = data.groupby(["Station","On or Off","Time"])["Station","Passenger"].mean()
```

```python
group_unstack = group.unstack(level=[1,2])
```

```python
group_unstack.head(21)
```

```python
group_unstack.describe()
```

```python
group_unstack.index
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(group_unstack)
```

```python
df_kmeans = pd.DataFrame(data = kmeans.labels_, index=group_unstack.index, columns= ['labels'])
```

```python
df_kmeans.head(21)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg = AgglomerativeClustering(n_clusters=3)
agg.fit(group_unstack)
```

```python
df_agg = pd.DataFrame(data = agg.labels_, index=group_unstack.index, columns= ['labels'])
```
