# 1. 시간별 유동인구 데이터 다운로드 및 전처리

```python
import pandas as pd
import numpy as np
```

* 3년치(2016년 2월 ~ 2019년 2월) 데이터를 다운받기 위해 datetime array 생성

```python
download_month = pd.period_range("2016-2", "2019-2", freq="M")
download_month
```

```python
gu_code = [26260, 26350]
```

* 문자열 format 함수로 url 문자열 생성

```python
example_day = int(''.join(str(download_month[0]).split('-')))
example_code = gu_code[1]
file_url = "https://bigdata.busan.go.kr/sgis/gis/getTimeExcelDownload.do?date={date}&gunguCd={code}&gunguTxt=&time=&timeTxt=%ec%a0%84%ec%b2%b4"
#file_url = "https://bigdata.busan.go.kr/sgis/gis/getTimeExcelDownload.do?date={date}&gunguCd=&gunguTxt=%ec%a0%84%ec%b2%b4&time=&timeTxt=%ec%a0%84%ec%b2%b4"
file_url.format(date=example_day, code=example_code)
```

```python
example_day
```

```python
example_code
```

* url에서 dataframe으로 읽어들이기

```python
example_df = pd.read_excel(file_url.format(date=example_day, code=example_code), header=2)
example_df.head(10)
```

```python
example_df.tail(20)
```

```python
example_df.shape
```

* 데이터 끝부분에 각 구별 월평균 이동량이 있으나, 시간대별 데이터에서 평균 column은 의미가 없는것 같아 drop 했습니다.

```python
example_df = example_df.drop(["평균"], axis=1)[:-18]
example_df.tail(10)
```

* 전치하여 시간축이 index, 각 구를 column으로 변환

```python
example_df = example_df.set_index("동별").T.iloc[1:]
example_df.head()
```

* 컬럼명이 일 수 만큼 중복되므로 rename 

```python
num_gu =len(set(example_df.columns))
gu = example_df.columns[:num_gu]
days = len(example_df.columns)/num_gu
new_columns  =[np.array(gu) +str(i) for i in range(1,int(days)+1)]
new_columns = np.concatenate(new_columns, axis=0)
new_columns
```

```python
example_df.columns = new_columns
example_df.head()
```

* 일별 데이터가 횡방향으로 있으므로 wide_to_long 함수로 dataframe을 reshape

```python
example_df['time'] = example_df.index
example_df = pd.wide_to_long(example_df, stubnames=gu, i='time',j='day')
example_df.head()
```

* time, day을 datetime로 parshing하여 index로 지정하고, 불필요한 column('time','day','계')은 drop

```python
example_df = example_df.reset_index()
example_df.index = pd.to_datetime(str(download_month[0])+'-'+example_df["day"].map(str)+'-'+example_df["time"], format ="%Y-%m-%d-%H시")
example_df = example_df.drop(['time','day','계'], axis =1)
example_df.head(20)
```

```python
example_df.tail(10)
```

* 전체 데이터 다운로드 및 전처리

```python
df_dong = []
df_hae = []
for month in download_month:
    for gu in gu_code:
        month_int = int(''.join(str(month).split('-')))
        df = pd.read_excel(file_url.format(date=month_int, code=gu), header=2)
        
        if gu == 26260:
            df = df.drop(["평균"], axis=1)[:-13]
        elif gu == 26350:
            df = df.drop(["평균"], axis=1)[:-18]
        df = df.set_index("동별").T.iloc[1:]

        num_dong =len(set(df.columns))
        dong = df.columns[:num_dong]
        days = len(df.columns)/num_dong
        new_columns  =[np.array(dong) +str(i) for i in range(1,int(days)+1)]
        new_columns = np.concatenate(new_columns, axis=0)
        df.columns = new_columns

        df['time'] = df.index
        df = pd.wide_to_long(df, stubnames=dong, i='time',j='day')

        df = df.reset_index()
        df.index = pd.to_datetime(str(month)+'-'+df["day"].map(str)+'-'+df["time"], format ="%Y-%m-%d-%H시")
        df = df.drop(['time','day','계'], axis =1)

        #df_list.append(df)
        
        if gu == 26260:
            df_dong.append(df)
        elif gu == 26350:
            df_hae.append(df)
            
        print(month_int)
        print(gu)
```

```python
df_total_dong = pd.concat(df_dong, axis = 0)
df_total_hae = pd.concat(df_hae, axis = 0)
data = pd.merge(df_total_dong, df_total_hae, left_index = True, right_index=True, how = "left")
#data = pd.concat(df_list, axis=0)
data
```

```python
data.to_csv('./data.csv')
```
