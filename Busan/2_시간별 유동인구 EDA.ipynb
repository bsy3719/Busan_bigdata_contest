* Boxplot 및 요약 통계량을 기준으로 유동인구가 많은 주요 6개 구를 선정

```python
major_gu = ['우1동', '우2동', '중1동', '온천1동', '온천2동', '재송1동' ]
```

```python
data['datetime-year'] = data.index.year
data['datetime-month'] = data.index.month
data['datetime-day'] = data.index.day
data['datetime-hour'] = data.index.hour
data['datetime-dayofweek'] = data.index.day_name()

time_columns = ['datetime-year', 'datetime-month', 'datetime-day', 'datetime-hour', 'datetime-dayofweek']
data[major_gu + time_columns].head()
```

* ## 년도별 주요구 유동인구

```python
fig, ((ax1, ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(nrows=2, ncols=3)
axes = [ax1, ax2, ax3, ax4, ax5, ax6]
fig.set_size_inches(32,12)
fig.suptitle('년도별 주요동 유동인구', fontsize=30)

for (gu, ax) in zip(major_gu, axes):
    ax.set_ylim(top=100000)
    ax.yaxis.label.set_visible(False)
    ax.set_title(gu, fontdict={'fontsize':20})
    sns.barplot(data=data, x='datetime-year', y=gu, ax = ax)
```

* ## 월별 주요구 유동인구

```python
fig, ((ax1, ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(nrows=2, ncols=3)
axes = [ax1, ax2, ax3, ax4, ax5, ax6]
fig.set_size_inches(32,12)
fig.suptitle('월별 주요동 유동인구', fontsize=30)

for (gu, ax) in zip(major_gu, axes):
    ax.set_ylim(top=100000)
    ax.yaxis.label.set_visible(False)
    ax.set_title(gu, fontdict={'fontsize':20})
    sns.barplot(data=data, x='datetime-month', y=gu, ax = ax)
```

* ## 시간별 주요구 유동인구

```python
fig, ((ax1, ax2,ax3),(ax4,ax5,ax6)) = plt.subplots(nrows=2, ncols=3)
axes = [ax1, ax2, ax3, ax4, ax5, ax6]
fig.set_size_inches(32,12)
fig.suptitle('시간별 주요동 유동인구', fontsize=30)

for (gu, ax) in zip(major_gu, axes):
    ax.set_ylim(top=100000)
    ax.yaxis.label.set_visible(False)
    ax.set_title(gu, fontdict={'fontsize':20})
    sns.barplot(data=data, x='datetime-hour', y=gu, ax = ax)
```

* 부산진구와 해운대구의 경우 새벽 2~3시 쯤의 유동인구가 많은 것으로 확인
* 데이터 측정 방식에 대한 의구심이 들어, 공단이 있어 출퇴근 유동인구가 명확할 것으로 보이는 강서구를 그래프로 확인
* 강서구의 경우 출근때 유동인구가 가장 많으며, 퇴근인 오후 6시 이후에 유동인구가 감소하는 것으로 확인

```python
plt.figure(figsize=(8,6))
sns.barplot(data=data, x='datetime-hour', y='우2동')
```

* ## 전체구 상관성 분석

```python
corr_matrix = data.iloc[:,:-5].corr(method='pearson')
plt.figure(figsize=(16,16))
sns.heatmap(corr_matrix, mask=np.zeros_like(corr_matrix, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True), square=True)
```
